```python
# Cual es la selecciÃ³n que mas penales metiÃ³?
# Completar
df_penalties = mundial[["1", "2", "1_panelties_scored", "2_panelties_scored"]]
selecciones_penalties = dict.fromkeys(equipos_total.unique(), 0)

for _, entry in df_penalties.iterrows():
  selecciones_penalties[entry["1"]] += entry["1_panelties_scored"]
  selecciones_penalties[entry["2"]] += entry["2_panelties_scored"]

max(selecciones_penalties, key=selecciones_penalties.get)
```

```python
# Cual es la selecciÃ³n que mas penales le hicieron?
# Completar

selecciones_pen_contra = dict.fromkeys(equipos_total.unique(), 0)

for _, entry in df_penalties.iterrows():
  selecciones_pen_contra[entry["1"]] += entry["2_panelties_scored"]
  selecciones_pen_contra[entry["2"]] += entry["1_panelties_scored"]

max(selecciones_pen_contra, key=selecciones_pen_contra.get)
```
